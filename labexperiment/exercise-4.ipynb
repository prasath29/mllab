```python
import numpy as np
import pandas as pd
```

```python
input_value=np.array([[0,0],[0,1],[1,1],[1,0]])
input_value.shape
input_value
```

```python
output=np.array([0,1,0,1])
output=output.reshape(4,1)
output.shape
```

```python
weights=np.array([[1.2],[1.2]])
weights
```

```python
bias = 0.5
```

```python
def sigmoid_func(x):
    return 1/(1+np.exp(-x))
```

```python
def der(x):
    return sigmoid_func(x)*(1-sigmoid_func(x))
```

```python
for epochs in range(15000):
    input_arr = input_value
    
    weighted_sum=np.dot(input_arr,weights)+bias
    first_output=sigmoid_func(weighted_sum)
```

```python
error=first_output - output
print(error)
```

```python
total_error=np.square(error).mean()
print(total_error)
```

```python
first_der=error
second_der=der(first_output)
derivative=first_der*second_der
print(second_der)
```

```python

print(derivative)
```

```python
t_input = input_value.T
print(t_input)
```

```python
final_derivative=np.dot(t_input,derivative)
print(final_derivative)
```

```python
for i in derivative:
        bias=bias-(0.05*i)
```
