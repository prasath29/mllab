```python
import numpy as np
import pandas as pd
```

```python
a = pd.read_csv("iris.csv")
print(a)
```

```python
a.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a.shape
```

```python
#univariate

a_setosa=a.loc[a['variety']=='Setosa']
a_setosa
```

```python
a_virginica=a.loc[a['variety']=='Virginica']
a_virginica
```

```python
a_versicolor=a.loc[a['variety']=='Versicolor']
a_versicolor
```

```python
plt.scatter(a_setosa['sepal.width'],np.zeros_like(a_setosa['sepal.width']))
plt.xlabel('sepal.width')
plt.show()
```

```python
plt.scatter(a_setosa['sepal.width'],np.zeros_like(a_setosa['sepal.width']))

plt.scatter(a_virginica['sepal.width'],np.zeros_like(a_virginica['sepal.width']))

plt.scatter(a_versicolor['sepal.width'],np.zeros_like(a_versicolor['sepal.width']))
plt.xlabel('sepal.width')
plt.show()
```

```python
plt.scatter(a_setosa['sepal.length'],np.zeros_like(a_setosa['sepal.length']),color="purple")

plt.scatter(a_virginica['sepal.length'],np.zeros_like(a_virginica['sepal.length']),color="red")

plt.scatter(a_versicolor['sepal.length'],np.zeros_like(a_versicolor['sepal.length']),color="blue")
plt.xlabel('sepal.length')
plt.show()
```

```python
plt.scatter(a_setosa['petal.length'],np.zeros_like(a_setosa['petal.length']),color="purple")

plt.scatter(a_virginica['petal.length'],np.zeros_like(a_virginica['petal.length']),color="red")

plt.scatter(a_versicolor['petal.length'],np.zeros_like(a_versicolor['petal.length']),color="blue")
plt.xlabel('petal.length')
plt.show()
```

```python
plt.scatter(a_setosa['petal.width'],np.zeros_like(a_setosa['petal.width']),color="purple")

plt.scatter(a_virginica['petal.width'],np.zeros_like(a_virginica['petal.width']),color="red")

plt.scatter(a_versicolor['petal.width'],np.zeros_like(a_versicolor['petal.width']),color="blue")
plt.xlabel('petal.width')
plt.show()
```

```python
# Bivariate regression

sns.FacetGrid(a,hue='variety',height=5).map(plt.scatter,"sepal.length","sepal.width").add_legend();
plt.show()
```

```python
sns.FacetGrid(a,hue='variety',height=5).map(plt.scatter,"petal.length","petal.width").add_legend();
plt.show()
```

```python
#Multivariate

sns.pairplot(a,hue="variety",height=2)
```
